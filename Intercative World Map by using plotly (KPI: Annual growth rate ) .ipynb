```python
data = pd.read_excel('/content/drive/MyDrive/Population Growth.xlsx')
```

```python
data.head(1)
```

```python
df=data.melt(id_vars=["Country Name", "Indicator Name"], 
        var_name="Year", 
        value_name="Value")
```

```python
df.to_excel(r'/content/drive/MyDrive/Data_student_project_map.xls')
```

```python
df
```

```python
dfff= df[df["Indicator Name"] == "Population growth (annual %)"]
df_countries= dfff[["Country Name","Year","Value"]]
```

```python
df_countries
```

```python
df_countries = df.groupby(['Country Name', 'Year']).sum().reset_index().sort_values('Year', ascending=False)
```

```python
start_date = df_countries["Year"].min()
end_date = df_countries["Year"].max()
mask = (df_countries['Year'] > start_date) & (df_countries['Year'] <= end_date)

df_countries = df_countries.loc[mask]

```

```python
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
df_countries
```

```python
fig = go.Figure(data=go.Choropleth(
locations = df_countries['Country Name'],
locationmode = 'country names',
z = df_countries['Value'],
colorscale = 'Reds',
marker_line_color = 'black',
marker_line_width = 0.5,
))

fig.update_layout(
title_text = 'Growth population rate',
title_x = 0.5,
geo=dict(
showframe = False,
showcoastlines = False,
projection_type = 'equirectangular'
)
)

fig.show()
```

```python


df_countrydate = df_countries.groupby(['Year','Country Name']).sum().reset_index()

#Creating the visualization
fig = px.choropleth(df_countrydate,
locations="Country Name",
locationmode = "country names",
color="Value",
hover_name="Country Name",
animation_frame="Year"
)

fig.update_layout(
title_text = 'Average Grwoth rate',
title_x = 0.5,
geo=dict(
showframe = False,
showcoastlines = False,
))

fig.show() 
```
